```julia
using Random
using CSV
using DelimitedFiles
using DataFrames
using Plots
#import DataFramesMeta as DFM
```

```julia
features_csv = "tmp/fix_features.csv"
data, header = readdlm(features_csv, ',', header=true)
df = DataFrame(data, vec(header))
#select!(df, Not(:NumberOfTurns))
```

```julia
using CatanLearning

for (name,feat) in CatanLearning.feature_library
    println(name)
    if string(name) in names(df)
        to_type = feat.type
        if feat.type == Bool
            to_type = Int8
        end
        df[!,name] = convert(Vector{to_type}, df[!,name])
    end
end
df
```

```julia

CSV.write(features_csv, df)
```
